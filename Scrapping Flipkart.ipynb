```python
import pandas as pd
import requests

# Set headers
headers = requests.utils.default_headers()
headers.update({ 'User-Agent': 'Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:52.0) Gecko/20100101 Firefox/52.0'})
from bs4 import BeautifulSoup

page = 1

product_name = []

product_price =[]

no_pages = int(input("Enter The Number of pages to scrap :"))
    
while page < no_pages:
    
    strings = "https://www.flipkart.com/search?sid=tyy%2C4io&otracker=CLP_Filters&p%5B%5D=facets.type%255B%255D%3DSmartphones&p%5B%5D=facets.rating%255B%255D%3D1%25E2%2598%2585%2B%2526%2Babove&page=" + str(page)

    req = requests.get(strings, headers)

    soup = BeautifulSoup(req.content, 'html.parser')

    a = soup.prettify

    name = soup.find_all( "div" ,class_ = '_3wU53n')
    
    price = soup.find_all(class_ ="_1vC4OE _2rQ-NK")
    
    for each in name:

        product_name.append(each.text) 
        
    for one in price:
        
        product_price.append(one.text)
        
        
    page +=1

print(len(product_price))
print(len(product_name))
```

```python
data = {"Product Name" : product_name,"Product Price" : product_price}

df = pd.DataFrame(data)

df
```
